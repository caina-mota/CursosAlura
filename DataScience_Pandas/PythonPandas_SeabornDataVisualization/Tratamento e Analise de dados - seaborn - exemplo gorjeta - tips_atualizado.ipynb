## Teste de hipÃ³tese

**H<sup>null</sup>**
> **A distribuiÃ§Ã£o da taxa da gorjeta Ã© a mesma nos dois grupos**

**H<sup>alt</sup>**
>**A distribuiÃ§Ã£o da taxa de gorjeta nÃ£o Ã© a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesas = gorjetas.query("sobremesa == 'Sim'").percentual
```

```python
sem_sobremesas = gorjetas.query("sobremesa == 'NÃ£o'").percentual
```

```python
sem_sobremesas
```

```python
r = ranksums(sobremesas, sem_sobremesas)
```

```python
print(f'O valor do p-value Ã© {r.pvalue}')
```

### ConclusÃ£o
**H<sup>alt</sup>**
>**A distribuiÃ§Ã£o da taxa de gorjeta nÃ£o Ã© a mesma nos dois grupos**
