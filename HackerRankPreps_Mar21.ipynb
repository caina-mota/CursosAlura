```python
def jumpingOnClouds(c):
    jumps = 0
    step = 0
    for i in range(len(c)-1):
        if i == step and (step+2)<len(c):
            if c[step+2] == 0:
                jumps = jumps+1
                step = step+2
            elif c[step+1] == 0:
                jumps = jumps+1
                step = step+1
        elif i == step and (step+1)<len(c):
            if c[step+1] == 0:
                jumps = jumps+1
                step = step+1
    return jumps
```

```python
def hourglassSum(arr):
    hourglass_sum = []
    for i in range(4):
        for j in range(4):
            h_sum = sum(arr[i][j:j+3]) + arr[i+1][j+1] + sum(arr[i+2][j:j+3])
            hourglass_sum.append(h_sum)
    max_sum = max(hourglass_sum)
    return max_sum
```

```python
def minimumBribes(q):
    bribes = []

    for queue_position, queue_value in enumerate(q):
        if ((queue_value-1)-queue_position) > 2:
            bribes.append('Too chaotic')
            break;
            #max(max(0,queue_value-2), queue_position)  para criar uma lista que pode conter -1 ou -2 devido ao valor -2, assim garantindovalor sepre iniciando em 0
        for i in range(max(0,queue_value-2), queue_position):
            if q[i] > queue_value:
                bribes.append(1)
    

    if 'Too chaotic' in bribes:
        output = 'Too chaotic'
    else:
        output = sum(bribes)
    
    print(output)
```
